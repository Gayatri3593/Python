```python
# print Odd Numbers in the list
NumList1 = (20,10,16,19,25,1,276,188)
for num in NumList1:
    if (num % 2 != 0):
       print(num, "is odd")
```

```python
# print Even numbers in the list
NumList1 = (20,10,16,19,25,1,276,188)
for num in NumList1:
    if (num % 2 == 0):
       print(num, "is even")
```
