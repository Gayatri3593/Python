```python
class OddEven:
    def OddEven():
        num = int(input("Enter a number:"))
        if(num%2==0):
            print(num," is Even number")
        else:
            print(num," is Odd number")

OddEven.OddEven()
```

```python
class EligiblityForMarriage:
    def Eligible():
        gender = input("Enter your Gender:")
        age = int(input("Enter your Age:"))
        if(gender == 'Male'):
            if(age >=21 ):
                print("ELIGIBLE")
            else:
                print("NOT ELIGIBLE")
        elif(gender == 'Female'):
            if(age >=18 ):
                print("ELIGIBLE")
            else:
                print("NOT ELIGIBLE")
EligiblityForMarriage.Eligible()
```
